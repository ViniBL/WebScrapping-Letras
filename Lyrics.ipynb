```python
from bs4 import BeautifulSoup
import requests
```

```python
def lyrics_extractor(html):
    print("1")
    soup = BeautifulSoup(html,'lxml')
    letra = soup.find_all('p')
    tamanho = len(letra)
    j = 0
    with open('letra.txt','a',encoding="utf-8") as f:
        f.write("\n\n")
        for i in letra:
            j+= 1
            if (j == tamanho-2):
                break
            else:
                f.write(i.get_text(strip=True,separator='\n'))
                f.write("\n\n")
```

```python
letras = requests.get('https://www.letras.mus.br/mais-acessadas/funk/').text
soup = BeautifulSoup(letras,'lxml')
links = soup.find('ol',class_="top-list_mus cnt-list--col1-3")

for link in links.find_all('a'):
    url = "https://www.letras.mus.br" + link.get("href")
    html = requests.get(url).text
    lyrics_extractor(html)



    
```

```python
import os
import re
import string
import random
from graph import Graph, Vertex
```

```python
def get_words_from_text(text_path):
    with open(text_path, 'rb') as file:
        text = file.read().decode("utf-8") 

        # remove [verse 1: artist]
        # include the following line if you are doing song lyrics
        # text = re.sub(r'\[(.+)\]', ' ', text)

        text = ' '.join(text.split())
        text = text.lower()
        text = text.translate(str.maketrans('', '', string.punctuation))

    words = text.split()

    words = words[:1000]

    return words
```

```python
def make_graph(words):
    g = Graph()
    prev_word = None
    # for each word
    for word in words:
        # check that word is in graph, and if not then add it
        word_vertex = g.get_vertex(word)

        # if there was a previous word, then add an edge if does not exist
        # if exists, increment weight by 1
        if prev_word:  # prev word should be a Vertex
            # check if edge exists from previous word to current word
            prev_word.increment_edge(word_vertex)

        prev_word = word_vertex

    g.generate_probability_mappings()
    
    return g
```

```python
def compose(g, words, length=50):
    composition = []
    word = g.get_vertex(random.choice(words))
    for _ in range(length):
        composition.append(word.value)
        word = g.get_next_word(word)

    return composition
```

```python
words = get_words_from_text('letra.txt')

# for song in os.listdir('songs/{}'.format(artist)):
    # if song == '.DS_Store':
    #     continue
    # words.extend(get_words_from_text('songs/{artist}/{song}'.format(artist=artist, song=song)))

g = make_graph(words)
composition = compose(g, words, 300)
print(' '.join(composition))
```
